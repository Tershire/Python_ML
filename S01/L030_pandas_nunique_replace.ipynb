```python
import numpy as np
import pandas as pd
```

```python
titanic_df = pd.read_csv("../data/titanic_train.csv")
# titanic_df = pd.read_csv("../data/titanic_train.csv", sep=",")
display(titanic_df)
```

```python
names = ["Dove", "Eagle", "Hawk", "Sparrow", "Eagle", "Kiwi"]
# scores = np.random.randint(low=0, high=100, size=len(names))
scores = np.array([97, None, 23, 23, 76, None])
passes = np.zeros_like(scores)
for i, score in enumerate(scores):
    if score is None:
        passes[i] = False
        
    elif score >= 60:
        passes[i] = True        
    else:
        passes[i] = False
    
df = pd.DataFrame(names, columns=["Name"])
df["Score"] = scores
df["Passed"] = passes

df["Note"] = df["Score"] / 100 * 20

display(df)
```

**.replace()**

```python
df["Passed"] = df["Passed"].replace(True, "T")

df["Passed"] = df["Passed"].replace({True: "T", False: "F"})
display(df)
```

*application as an alternative to .fillna()*

```python
column_names = ["Score", "Note"]

df[column_names] = df[column_names].replace(np.nan, 0)
display(df)
```
