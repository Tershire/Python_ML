```python
import numpy as np
import pandas as pd
```

```python
titanic_df = pd.read_csv("./data/titanic_train.csv")
# titanic_df = pd.read_csv("./data/titanic_train.csv", sep=",")
display(titanic_df)
```

```python
scores = np.random.randint(low=0, high=100, size=10)
passes = scores > 60

df = pd.DataFrame(scores, columns=["Score"])
df["Passed"] = passes

display(df)
```

**get indices**

```python
indices = df.index

display(indices)
print(indices.values)
```

*pick some*

```python
picks = [1, 3, 9]
picks = np.arange(1, 10, 2)

print(indices[picks].values) # pandas indices can be directly indexed
print(indices.values[picks])

print(indices[7])
```

*test opeerations on series*

```python
score_sr = df["Score"]
print(score_sr)

print(score_sr.max(), max(score_sr))
print(score_sr.sum(), sum(score_sr))
```

**reset_index()**

```python
df_index_renewed = df.reset_index(inplace=False)

display(df_index_renewed)
```

*example: .value_counts()*

```python
Pclass_sr = titanic_df["Pclass"]
Pclass_counts = Pclass_sr.value_counts()
print(Pclass_counts, type(Pclass_counts))

Pclass_counts_1 = Pclass_counts.reset_index(inplace=False)
display(Pclass_counts_1, type(Pclass_counts_1)) # note data type is DataFrame, not Series

Pclass_counts_2 = Pclass_counts.reset_index(drop=True, inplace=False)
display(Pclass_counts_2, type(Pclass_counts_2)) # note data type is Series
```

*common use*

```python
Pclass_counts = titanic_df["Pclass"].value_counts().reset_index()
Pclass_counts.rename(columns={"index":"Pclass #", "Pclass":"Pclass counts"}, inplace=True)

display(Pclass_counts)
```
