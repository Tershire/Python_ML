```python
import numpy as np
import pandas as pd
```

```python
titanic_df = pd.read_csv("../data/titanic_train.csv")
# titanic_df = pd.read_csv("../data/titanic_train.csv", sep=",")
display(titanic_df)
```

```python
names = ["Dove", "Eagle", "Hawk", "Sparrow", "Eagle", "Kiwi"]
# scores = np.random.randint(low=0, high=100, size=len(names))
scores = np.array([97, None, 23, 23, 76, None])
passes = np.zeros_like(scores)
for i, score in enumerate(scores):
    if score is None:
        passes[i] = False
        
    elif score >= 60:
        passes[i] = True        
    else:
        passes[i] = False
    
df = pd.DataFrame(names, columns=["Name"])
df["Score"] = scores
df["Passed"] = passes

df["Note"] = df["Score"] / 100 * 20

display(df)
```

**.isna()**

```python
display(df.isna())
```

```python
# count number of na s 
print(df.isna().sum())
```

**.fillna()**

```python
print(df["Score"].fillna(0))
print(df["Score"].fillna(df["Score"].mean()))

df["Score"] = df["Score"].fillna(df["Score"].mean())
display(df)
print(df.isna().sum())
```

```python
# fill na: Age
titanic_df["Age"] = titanic_df["Age"].fillna(titanic_df["Age"].mean())

# fill na: Cabin
titanic_df["Cabin"] = titanic_df["Cabin"].fillna("C###")

# fill na: Embarked
titanic_df["Embarked"] = titanic_df["Embarked"].fillna("S")

# check
print(titanic_df.isna().sum())
display(titanic_df)
```
