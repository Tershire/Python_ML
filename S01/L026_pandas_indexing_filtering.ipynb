```python
import numpy as np
import pandas as pd
```

```python
titanic_df = pd.read_csv("../data/titanic_train.csv")
# titanic_df = pd.read_csv("../data/titanic_train.csv", sep=",")
display(titanic_df)
```

```python
names = ["Eagle", "Penguin", "Hawk", "Sparrow", "Kiwi"]
scores = np.random.randint(low=0, high=100, size=len(names))
passes = scores > 60

df = pd.DataFrame(names, columns=["Name"])
df["Score"] = scores
df["Passed"] = passes

df["Note"] = df["Score"] / 100 * 20

display(df)
```

*DataFrame[str] -> Series*

```python
score_sr = df["Score"].head(3)
display(score_sr, type(score_sr), score_sr.shape)
```

*DataFrame[list] -> DataFrame*

```python
# single column
score_df = df[["Score"]].head(3)
display(score_df, type(score_df), score_df.shape)

# multiple columns
sub_df = df[["Score", "Note"]].head(3)
display(sub_df, type(sub_df), sub_df.shape)
```

```python
df.index = ["I", "II", "III", "IV", "V"]

display(df)
```

**iloc[]**

```python
print(df.iloc[1, 0])
```

```python
# all
display(df.iloc[:])

# 
display(df.iloc[1:4])

# 
display(df.iloc[[0, 2, 3], :-1])
```

*<!> iloc[] does not support Boolean Indexing <!>*

**loc[]**

```python
print(df.loc["II", "Name"])
```

```python
# 
display(df.loc["II":"IV"])

# 
display(df.loc[["I", "III", "IV"], ["Name", "Score", "Passed"]])
```

*<O> loc[] supports Boolean Indexing <O>*

```python
display(df.loc[df.Passed])

display(df.loc[df.Score % 2 == 1])
```
