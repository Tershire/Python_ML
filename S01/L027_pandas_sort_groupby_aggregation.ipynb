```python
import numpy as np
import pandas as pd
```

```python
titanic_df = pd.read_csv("../data/titanic_train.csv")
# titanic_df = pd.read_csv("../data/titanic_train.csv", sep=",")
display(titanic_df)
```

```python
names = ["Eagle", "Eagle", "Hawk", "Sparrow", "Kiwi", "Eagle"]
# scores = np.random.randint(low=0, high=100, size=len(names))
scores = np.array([96, 23, 97, 23, None, 67])
passes = np.zeros_like(scores)
for i, score in enumerate(scores):
    if score is None:
        passes[i] = False
        
    elif score >= 60:
        passes[i] = True        
    else:
        passes[i] = False
    
df = pd.DataFrame(names, columns=["Name"])
df["Score"] = scores
df["Passed"] = passes

df["Note"] = df["Score"] / 100 * 20

display(df)
```

**.sort_values()**

```python
# ascending (default)
sorted_df = df.sort_values(by=["Name"])

display(sorted_df)

# descending
sorted_df = df.sort_values(by=["Name"], ascending=False)

display(sorted_df)
```

*sort with multiple priotized critieria*

```python
sorted_df = df.sort_values(by=["Name", "Score"])

display(sorted_df)
```

**aggregation functions**

*.count()*

```python
print(df.count())  # None or NaN are not counted
```

```python
column_names = ["Name", "Score"]

# .count()
print(df[column_names].count())

# .sum()
print(df[column_names].sum())

# .mean()
print(df["Score"].mean())
```

```python
print(df["Score"].agg([sum, max]))
```

**.groupby()**

```python
dfgb = df.groupby(by="Name")

display(dfgb)
```

```python
dfgb.mean()
```

```python
dfgb[["Score", "Note"]]

dfgb[["Score", "Note"]].mean()
```

*Quick Review*

```python
display(df.groupby(by="Score").count())

display(df.groupby(by="Score")[["Name", "Passed"]].count())
```

**.agg()**

```python
display(df.groupby(by="Score")[["Name", "Passed"]].agg([sum, max]))

display(df.groupby(by="Name")[["Score", "Passed"]].agg([sum, "mean"]))
```

```python
agg_format = {"Score": sum, "Passed": "mean"}

display(df.groupby(by=["Name"])[["Score", "Passed"]].agg(agg_format))
```

*Ex) Titanic Data*

```python
agg_format = {"Age": max, "SibSp": sum, "Fare": "mean"}

display(titanic_df.groupby(by=["Pclass"]).agg(agg_format))
```

```python
agg_format = {"Age": [max, "mean"], "Fare": "mean"}

display(titanic_df.groupby(by=["Pclass"]).agg(agg_format))
```

**named aggregation (NamedAgg)**

```python
titanic_df.groupby(by=["Pclass"]).agg(age_max=("Age", max), age_mean=("Age", np.mean), fare_mean=("Fare", "mean"))
```

```python
titanic_df.groupby(by=["Pclass"]).agg(
    age_max=pd.NamedAgg(column="Age", aggfunc=max),
    age_mean=pd.NamedAgg(column="Age", aggfunc=np.mean),
    fare_mean=pd.NamedAgg(column="Fare", aggfunc="mean")
)
```
